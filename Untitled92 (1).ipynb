```python
import pandas
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pandas.read_csv("/content/open.csv")
```

```python
data.shape
data['Indicator Code'].unique()
```

```python
data.describe()
```

```python
awb=data[data['Country Code']=='ABW']
# Filter the data for the AUS and GDP per capita of 20 years 
plt.figure(figsize=(30,8))
awb=awb[:20]
# Create a line plot
plt.plot(awb.iloc[:, 4:].columns, awb.iloc[:, 4:].values[0])
plt.title('GDP per capita of the ABW')
plt.xlabel('Year')
plt.ylabel('GDP per capita')
plt.show()
```

```python
# Filter the data for the AUS and GDP per capita of 20 years 
plt.figure(figsize=(30,8))
aus_data = data[(data['Country Code'] == 'AUS')]
aus_data=aus_data[:20]
# Create a line plot
plt.plot(aus_data.iloc[:, 4:].columns, aus_data.iloc[:, 4:].values[0])
plt.title('GDP per capita of the AUS')
plt.xlabel('Year')
plt.ylabel('GDP per capita')
plt.show()
```
